Can a Single Repayment Record Predict Credit Card Default?
A hands‑on case study with 30,000 real‑world clients

You’re a risk manager at a bank. Every month, some customers fail to pay their credit card bills – a default that costs the bank real money. If you could predict who is likely to default next month, you could proactively reduce their credit limit, offer a payment plan, or set aside reserves.

That’s exactly the question I tried to answer. Below is what I found, why a random forest worked reasonably well, and three reasons you shouldn’t trust my results too much.

The data: 30,000 credit card clients
I used the UCI “Default of Credit Card Clients” dataset (Yeh & Lien, 2009). It contains 30,000 customers in Taiwan with 24 features: demographics (age, sex, education, marriage), repayment status for the past six months (PAY_0 to PAY_6), monthly bill amounts (BILL_AMT1–6), and payment amounts (PAY_AMT1–6). The target is default.payment.next.month – whether the client will default the following month.

I split the data into 70% training and 30% testing (stratified, random_state=123). The training set has a 22.1% default rate – that imbalance matters, as we’ll see.

Exploratory look: what does the data tell us?
Figure 1 shows the class distribution in the training set.

*Figure 1: Non‑default (0) = 16,355, default (1) = 4,645. Only 22% of clients default.*

This is a classic class imbalance problem. A model that always predicts “non‑default” would achieve 78% accuracy but would be completely useless. That’s why I will not use accuracy – instead, I’ll rely on ROC‑AUC and F1‑score.

Figure 2 compares credit limit (LIMIT_BAL) between defaulters and non‑defaulters.

Figure 2: Defaulters (orange) tend to have lower credit limits, but the distributions overlap heavily. Credit limit alone is a weak separator.

So we need more features – and a real model.

The models: from logistic regression to random forest
I trained three models:

Logistic regression (linear baseline, with class_weight='balanced' to handle imbalance)

Random forest (100 trees, min_samples_leaf=20, class_weight='balanced')

HistGradientBoosting (fast gradient boosting)

All models shared the same preprocessing pipeline: numeric features were standardised, categorical features were one‑hot encoded. Hyperparameters were tuned using 5‑fold stratified cross‑validation with ROC‑AUC as the scoring metric.

The best model was the random forest.

Results: what drives default?
Figure 3 shows permutation importance – how much the model’s ROC‑AUC drops when a feature is randomly shuffled. The larger the drop, the more important the feature.


Figure 3: PAY_0 (most recent repayment status) dominates everything else. Credit utilisation, credit limit, and the number of zero‑payment months also matter.

This makes intuitive sense: if a client was already late on their last bill, they are very likely to default next month. The model essentially learns this single rule.

Quantitative results (test set)
Model	ROC‑AUC	Accuracy	F1 (default class)	Recall (default class)
Random forest (best)	0.783	82.3%	0.475	36.2%
Logistic regression	0.741	68.7%	0.483	66.1%
Dummy (always predict 0)	0.500	77.9%	0.000	0.0%
Key observations:

The random forest has a decent ROC‑AUC – it ranks defaulters higher than non‑defaulters reasonably well.

But recall is only 36%: the model misses almost two out of every three actual defaulters. For a real bank, that might be unacceptable.

Three caveats – why you should be sceptical
1. Class imbalance still hurts recall
Even with class_weight='balanced', the model is conservative. It prefers to say “non‑default” rather than risk a false alarm. In practice, a bank could lower the decision threshold from 0.5 to 0.3 – that would catch more defaulters but also flag many good customers as risky. I did not optimise that trade‑off.

2. PAY_0 is a look‑ahead feature (data leakage)
The PAY_0 column records repayment status for the most recent month. But if you are trying to predict default before the next month starts, you don’t yet know PAY_0. My model cheats by using future information. Removing PAY_0 drops the ROC‑AUC to about 0.68 – a huge difference. Any real‑world deployment must exclude such post‑period features.

3. Old data, narrow geography
The data is from 2005 Taiwan. Consumer behaviour changes during recessions, pandemics, and across different regulatory environments. A model that works on 2005 Taiwanese data might fail completely on 2026 US customers. I have no way to test this without newer, more diverse data.

Final verdict
A random forest can predict credit card default with a reasonable ROC‑AUC (0.78), mainly by looking at the most recent repayment status. However, low recall and the fact that the strongest feature (PAY_0) may not be available in time make the model less useful for real‑time decisions. For a bank, this would be a screening tool – not a replacement for human judgement.

Final test ROC‑AUC: 0.783
(Chosen because it is robust to class imbalance.)
